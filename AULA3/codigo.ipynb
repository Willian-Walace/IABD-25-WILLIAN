```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0, 10, 10)       
dados_y = dados_x ** 2
```

```python
plt.figure(figsize=(10, 8),facecolor='grey')

ax = plt.axes()
ax.set_facecolor('white')
plt.plot(dados_x, dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')

plt.xlabel('Eixo X', fontsize=16)
plt.ylabel('Eixo Y', fontsize=16)
plt.title('Gr√°fico simples com Matplotlib', fontsize=20)
```
