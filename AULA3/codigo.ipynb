```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0, 10, 10)       
dados_y = dados_x ** 2
```

```python
plt.figure(figsize=(10, 8),facecolor='grey')

ax = plt.axes()
ax.set_facecolor('white')
plt.plot(dados_x, dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')

plt.xlabel('Eixo X', fontsize=16)
plt.ylabel('Eixo Y', fontsize=16)
plt.title('Gr√°fico simples com Matplotlib', fontsize=20)
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(12, 12))
linha = 2 
coluna = 2
# Primeiro gr√°fico
plt.subplot(linha, coluna, 1)
plt.plot(dados_x,'r--',label='dados_x')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')       
plt.title('Gr√°fico 1')

# Segundo gr√°fico
plt.subplot(linha, coluna, 2)
plt.plot(dados_x,dados_y,'b--',label='dados_x\ne y')
plt.legend(loc='lower right')    
plt.xlabel('Eixo X')        
plt.ylabel('Eixo Y')
plt.title('Gr√°fico 2')

# Terceiro gr√°fico
plt.subplot(linha, coluna, 3)   
plt.plot(dados_x*dados_x,'g-',label='dados_x ^ 2')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')    
plt.ylabel('Eixo Y')
plt.title('Gr√°fico 3')

# Quarto gr√°fico
plt.subplot(linha, coluna, 4)
plt.plot(dados_y,'c--',label='dados_y')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gr√°fico 4')

plt.savefig('graficos_subplots.png')
plt.show
```

```python
figura = plt.figure()
grafico = figura.add_axes([0, 0, 0.7, 0.7])
rotulos = ['A', 'B', 'C', 'D', 'E', 'F','G', 'H', 'I', 'J']
grafico.bar(rotulos, dados_x, color='red')
grafico.set_xlabel('Rotulos')
grafico.set_ylabel('Qunatidade')
```
