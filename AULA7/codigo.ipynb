```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
vazao = np.array([10,20,30,5,35,40]).reshape(-1, 1)
nivel = np.array([50,60,70,45,80,85])

modelo = LinearRegression()
modelo.fit(vazao, nivel)

```

```python
#vazao_teste = np.array([15,25,33]).reshape(-1, 1)
vazao_teste = np.array([5,2,7,30,50]).reshape(-1, 1)

nivel_previsto = modelo.predict(vazao_teste)
print(nivel_previsto)
```

```python

for i in range(len(vazao_teste)):
    print(f'Para vazão de {vazao_teste[i][0]} m3/s, o nível previsto é {nivel_previsto[i]:.2f} m')

plt.scatter(vazao, nivel, color='blue', label='Dados Treinamento')
plt.plot(vazao_teste, nivel_previsto, color='red', label='Previsão')
plt.xlabel('Vazão (m³/s)')
plt.ylabel('Nível (m)')
plt.title('Previsão de Nível com Base na Vazão')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

```

```python
Q = np.array([100,200,300,400,500,600,700]).reshape(-1, 1)
T = np.array([30,35,40,45,60,100,150])
```

```python
print(Q)
print(T)
```

```python
linear_model = LinearRegression()
linear_model.fit(Q, T)
T_linear_pred = linear_model.predict(Q)
print(T_linear_pred)
```

```python
poly = PolynomialFeatures(degree=5)
Q_poly = poly.fit_transform(Q)
poly_model = LinearRegression()
poly_model.fit(Q_poly, T)
```

```python
Q_range = np.linspace(100, 700, 500).reshape(-1, 1)
T_linear_curve = linear_model.predict(Q_range)
t_poly_curve = poly_model.predict(poly.transform(Q_range))

plt.figure(figsize=(10, 6))
plt.scatter(Q, T, color='black', label='Dados Originais')
plt.plot(Q_range, T_linear_curve, color='green', label='Regressão Linear')
plt.plot(Q_range, t_poly_curve, color='blue', linestyle='--', label='Regressão Polinomial (Grau 2)')
plt.title('Comparação entre Regressão Linear e Polinomial')
plt.xlabel('Quantidade de calor (Q)')
plt.ylabel('Temperatura (T)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
