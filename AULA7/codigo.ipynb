```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
vazao = np.array([10,20,30,5,35,40]).reshape(-1, 1)
nivel = np.array([50,60,70,45,80,85])

modelo = LinearRegression()
modelo.fit(vazao, nivel)

```

```python
#vazao_teste = np.array([15,25,33]).reshape(-1, 1)
vazao_teste = np.array([5,2,7,30,50]).reshape(-1, 1)

nivel_previsto = modelo.predict(vazao_teste)
print(nivel_previsto)
```

```python

for i in range(len(vazao_teste)):
    print(f'Para vazão de {vazao_teste[i][0]} m3/s, o nível previsto é {nivel_previsto[i]:.2f} m')

plt.scatter(vazao, nivel, color='blue', label='Dados Treinamento')
plt.plot(vazao_teste, nivel_previsto, color='red', label='Previsão')
plt.xlabel('Vazão (m³/s)')
plt.ylabel('Nível (m)')
plt.title('Previsão de Nível com Base na Vazão')
plt.legend()
plt.grid()
plt.show()
```
