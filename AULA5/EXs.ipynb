```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Inteligencia_Artificial_2025/refs/heads/main/bases_de_dados/monitoramento_tempo.csv')
```

```python
import datetime 
df['data'] = pd.to_datetime(df['data'])
```

```python
fig = plt.figure(figsize=(12,5))
eixo = fig.add_axes([0,0,1,1])
eixo.xaxis.set_major_locator(mdates.DayLocator(interval=2))
eixo.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
eixo.set_ylim(0,100)
eixo.plot(df['data'], df['humidade'], color='b')
eixo.grid(False)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.tick_params(axis='x',rotation=45)
eixo.set_title('Umidade ao longo do tempo', fontsize=25,pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Umidade', fontsize=20)
eixo.legend()
```

```python
fig = plt.figure(figsize=(14,6))
eixo = fig.add_axes([0,0,1.1,1.1])
eixo.xaxis.set_major_locator(mdates.DayLocator(interval=2))
eixo.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
eixo.set_ylim(0,15)
eixo.plot(df['data'], df['velocidade do vento'], color='r')
eixo.grid(False)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.tick_params(axis='x',rotation=90)
eixo.set_title('velocidade do vento ao longo do tempo', fontsize=24,pad=19)
eixo.set_xlabel('Ano', fontsize=21)
eixo.set_ylabel('velocidade do vento', fontsize=21)
eixo.legend()
```

```python

# Criar figura
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['humidade'], color='b')
eixo.grid(True)
eixo.set_title('Umidade no momento', fontsize=25, pad=20)
eixo.set_ylabel('Umidade (%)', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['humidade'], loc='best')
eixo.set_xlim(datetime.datetime(2012,10,1), datetime.datetime(2014,6,1))


eixo.axhline(df['humidade'].max(), color='k', linestyle='--')
eixo.axhline(df['humidade'].min(), color='k', linestyle='--')

# Anotao do valor mximo
x1 = df['data'][df['humidade'].idxmax()]  # posio em X
y1 = max(df['humidade'])                   # valor mximo
x2 = df['data'][df['humidade'].idxmax()+7000] if df['humidade'].idxmax() < 7000 else df['data'][0]
y2 = max(df['humidade']) + 5               # desloca o texto para cima
eixo.annotate('Mximo', xy=(x1, y1), fontsize=20,
              xytext=(x2, y2), arrowprops=dict(facecolor='k'))

# Anotao do valor mnimo
x1 = df['data'][df['humidade'].idxmin()]  # posio em X
y1 = min(df['humidade'])                   # valor mnimo
x2 = df['data'][df['humidade'].idxmin()+7000] if df['humidade'].idxmin() < 7000 else df['data'][0]
y2 = min(df['humidade']) + 5               # desloca o texto para cima
eixo.annotate('Mnimo', xy=(x1, y1), fontsize=20,
              xytext=(x2, y2), arrowprops=dict(facecolor='k'))

plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
plt.rcParams['font.size'] = '10'
eixo = fig.add_axes([0,0,1,1])
eixo.set_ylim(0,15)
eixo2 = fig.add_axes([0.65,0.58,0.3,0.3])
eixo.plot(df['data'], df['velocidade do vento'], color='b')
eixo.grid(True)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.tick_params(axis='x',rotation=45)
eixo.set_title('velocidade do vento ao longo do tempo', fontsize=25,pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('velocidade do vento', fontsize=20)

eixo.legend(['velocidade do vento'],loc ='best',fontsize=10)
azul_esquerda = df['data']<datetime.datetime(2014,5,1)
azul_direita = df['data']>datetime.datetime(2014,6,1)

eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo2.plot(df['data'], df['velocidade do vento'], color='r')
eixo2.plot(df[azul_esquerda]['data'],df[azul_esquerda]['velocidade do vento'],color = 'b')
eixo2.plot(df[azul_direita]['data'],df[azul_direita]['velocidade do vento'],color = 'b')
eixo2.grid(True)
eixo.set_title('velocidade do vento ao longo do tempo', fontsize=25,pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('velocidade do vento', fontsize=20)
eixo2.legend(['velocidade do vento'],loc ='best',fontsize=10)
```

```python
fig = plt.figure(figsize=(15,5))
plt.rcParams['font.size'] = '10'
eixo = fig.add_axes([0,0,1,1])
eixo.set_ylim(950,1200)
eixo2 = fig.add_axes([0.65,0.58,0.3,0.3])
eixo.plot(df['data'], df['presso'], color='b')
eixo.grid(True)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.tick_params(axis='x',rotation=45)
eixo.set_title('presso ao longo do tempo', fontsize=25,pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('presso', fontsize=20)

eixo.legend(['presso'],loc ='best',fontsize=10)
azul_esquerda = df['data']<datetime.datetime(2014,5,1)
azul_direita = df['data']>datetime.datetime(2014,6,1)

eixo.axhline(max(df['presso']),color='k',linestyle='--')
eixo.axhline(min(df['presso']),color='k',linestyle='--')

# Anotao do valor mximo
x1 = df['data'][df['humidade'].idxmax()]  # posio em X
y1 = max(df['humidade'])                   # valor mximo
x2 = df['data'][df['humidade'].idxmax()+7000] if df['humidade'].idxmax() < 7000 else df['data'][0]
y2 = max(df['humidade']) - 5               # desloca o texto para cima
eixo.annotate('Mximo', xy=(x1, y1), fontsize=20,
              xytext=(x2, y2), arrowprops=dict(facecolor='k'))

# Anotao do valor mnimo
x1 = df['data'][df['humidade'].idxmin()]  # posio em X
y1 = min(df['humidade'])                   # valor mnimo
x2 = df['data'][df['humidade'].idxmin()+7000] if df['humidade'].idxmin() < 7000 else df['data'][0]
y2 = min(df['humidade']) + 5               # desloca o texto para cima
eixo.annotate('Mnimo', xy=(x1, y1), fontsize=20,
              xytext=(x2, y2), arrowprops=dict(facecolor='k'))




eixo2.plot(df['data'], df['presso'], color='r')
eixo2.plot(df[azul_esquerda]['data'],df[azul_esquerda]['presso'],color = 'b')
eixo2.plot(df[azul_direita]['data'],df[azul_direita]['presso'],color = 'b')
eixo2.grid(True)
eixo.set_title('presso ao longo do tempo', fontsize=25,pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('presso', fontsize=20)
eixo2.legend(['presso'],loc ='best',fontsize=10)


plt.show()
```
